## **Feature Engineering**

```python
# Dropping Unnamed column & id column
df = df.drop(['Unnamed: 32', 'id'], axis=1)
```

```python
df.columns
```

```python
# Checking unique values for diagnosis column
df['diagnosis'].unique()
```

```python
# Converting diagnosis column object to int
df['diagnosis'] = df['diagnosis'].map({'M' : 1, 'B' : 0})
```

```python
df['diagnosis'].unique()
```

```python
# Split data to X, y
X = df.drop(['diagnosis'], axis = 1)
y = df['diagnosis']
```

## **Feature Scaling**

```python
# Scaling my X data
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

## **Train & Test split**

```python
# Split data to train and test
X_test, X_train, y_test, y_train = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## **Model Training**
### **1. Logistic Regression**

```python
# Training
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_model_pred = log_model.predict(X_test)
log_accuracy = accuracy_score(y_test, log_model_pred)
print("Logistic Regression Accuracy:", log_accuracy)
print("Classification Report:\n", classification_report(y_test, log_model_pred))
```

**Accuracy**: 95.6% â€” Overall, your model correctly predicted 95.6% of the cases.

**1. Precision:**
- Class 0 (negative): 97% â€” When the model predicts class 0, it is correct 97% of the time.
- Class 1 (positive): 94% â€” When the model predicts class 1, it is correct 94% of the time.

**2. Recall:**
- Class 0 (negative): 97% â€” Of all the actual class 0 cases, the model correctly identified 97% of them.
- Class 1 (positive): 94% â€” Of all the actual class 1 cases, the model correctly identified 94% of them.

**3. F1-Score:**
- Class 0: 0.97 â€” A balance between precision and recall for class 0.
- Class 1: 0.94 â€” A balance between precision and recall for class 1.

### **2. Random Forest**

```python
rad_model = RandomForestClassifier()
rad_model.fit(X_train, y_train)
rad_model_pred = rad_model.predict(X_test)
rad_accuracy = accuracy_score(y_test, rad_model_pred)
print("Random Forest Accuracy:", rad_accuracy)
print("Classification Report:\n", classification_report(y_test,rad_model_pred))
```
